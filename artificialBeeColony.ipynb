```python
import numpy as np
import pandas as pd
from scipy.io import loadmat
import matplotlib.pyplot as plt

from google.colab import files
upload = files.upload()
data = loadmat('BSE30.mat')

expr=data['ExpReturn']
expcov=data['ExpCovariance']

df=pd.DataFrame(expr)
df.to_csv()
display(df)

df2=pd.DataFrame(expcov)
df2.to_csv()
display(df2)

Returns=df
covariance=df2
```

```python
def find_AR_i(Returns,covariance,weights):
  P_risk=np.sqrt(np.dot(weights.T,np.dot(covariance,weights)))
  MR=np.dot(covariance,weights)/P_risk
  AR=weights*MR
  return AR

def Risk_Budgeting_constraint(Returns,covariance,AR,weights):
  P_risk=np.sqrt(np.dot(weights.T,np.dot(covariance,weights)))
  dim=len(weights)
  for i in range(dim):
    if AR[i]>0.125*P_risk:
      return False
  return True

def roulette_wheel_selection(P):
    r = np.random.rand()
    C = np.cumsum(P)
    i = np.argmax(r <= C)  # Find the first index where r <= C
    return i

def sharpeRatio(Returns,covariance,unscaled_weights):
  weights=softmax(unscaled_weights)
  P_return=np.sum((Returns.mean()*weights))
  P_risk=np.sqrt(np.dot(weights.T,np.dot(covariance,weights)))
  # AR=find_AR_i(Returns,covariance,weights)
  # if Risk_Budgeting_constraint(Returns,covariance,AR,weights)==False:
  #    return 1
  return -P_return/P_risk

def isFeasible(Returns,covariance,unscaled_weights):
  dim=len(unscaled_weights)
  weights=softmax(unscaled_weights)
  AR=find_AR_i(Returns,covariance,weights)
  if Risk_Budgeting_constraint(Returns,covariance,AR,weights)==False:
    return False
  else:
    return True

def softmax(x):
    e_x = np.exp(x - np.max(x))
    return e_x / e_x.sum(axis=0)


def ABC(Returns,covariance,n_iters,n_popn):

  dim=len(Returns.columns)
  n_onlookers=n_popn
  L=np.round(0.6*dim*n_popn)#Abandonment Limit Parameter (Trial Limit)
  a=1  #Acceleration Coefficient Upper Bound

  individuals=np.zeros((n_popn,dim))
  fitness=np.zeros(n_popn)

  feasible=[]
  for i in range(n_popn):
    individuals[i,:]=np.random.rand(1,dim)
    fitness[i]=sharpeRatio(Returns,covariance,individuals[i,:])

  indF = np.argsort(fitness)[::-1]
  individuals = individuals[indF, :]
  fitness = fitness[indF]

  BestF = fitness[-1]
  BestX = individuals[-1, :]

  C=np.zeros((n_popn,1))#Abandonment Counter

  new_indi=np.zeros((n_popn,dim))

  history=[]

  for it in range(n_iters):

    #Recruited bees
    for i in range(n_popn):

      # Choose k randomly, not equal to i
      K = list(range(0,i))+list(range(i + 1,n_popn))
      k = np.random.choice(K)

      # Define Acceleration Coeff.
      phi = a * np.random.uniform(-1, 1,dim)

      new_indi[i,:]=individuals[i,:]+phi*(individuals[i,:]-individuals[k,:])

      new_fit=sharpeRatio(Returns,covariance,new_indi[i,:])

      if new_fit<fitness[i]:
        individuals[i,:]=new_indi[i,:]
        fitness[i]=new_fit
      else:
        C[i]=C[i]+1

    #Calculate Fitness Values and Selection Probabilities
    meancost=np.mean(fitness)
    F=np.zeros(n_popn)
    for i in range(n_popn):
      F[i]=np.exp(-fitness[i]/meancost)
    P=F/np.sum(F)

    #Onlooker bees
    for m in range(n_onlookers):

      i=roulette_wheel_selection(P)

      # Choose k randomly, not equal to i
      K = list(range(0,i))+list(range(i + 1,n_popn))
      k = np.random.choice(K)

      # Define Acceleration Coeff.
      phi = a * np.random.uniform(-1, 1,dim)

      new_indi[i,:]=individuals[i,:]+phi*(individuals[i,:]-individuals[k,:])

      new_fit=sharpeRatio(Returns,covariance,new_indi[i,:])

      if new_fit<fitness[i]:
        individuals[i,:]=new_indi[i,:]
        fitness[i]=new_fit
      else:
        C[i]=C[i]+1

    #Scout Bees
    for i in range(n_popn):
      if C[i]>=L:
        individuals[i,:]=np.random.uniform(0,1,dim)
        fitness[i]=sharpeRatio(Returns,covariance,individuals[i,:])
        C[i]=0


    indF = np.argsort(fitness)[::-1]
    individuals = individuals[indF,:]
    fitness = fitness[indF]

    BestF = fitness[-1]
    BestX = individuals[-1,:]
    print(softmax(BestX))
    print(-BestF)
    history.append(-BestF)


  return history,softmax(BestX),-BestF

history,BestX,Sharpe=ABC(Returns,covariance,100,100)
print("history =",history)
print("BestX",BestX)
print("Sharpe Ratio =",Sharpe)
```
