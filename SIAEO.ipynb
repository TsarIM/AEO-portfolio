```python
def sharpeRatio(Returns,covariance,unscaled_weights):

  weights=unscaled_weights/np.sum(unscaled_weights)

  P_return=np.sum((Returns.mean()*weights))

  P_risk=np.sqrt(np.dot(weights.T,np.dot(covariance,weights)))

  return -P_return/P_risk
```

```python
def spaceBound(X,lb,ub):
  for i in range(len(X)):
    if X[i]<lb[i] or X[i]>ub[i]:
      X[i]=np.random.uniform(lb[i],ub[i])
  return X
```

```python
def SIAEO(Returns,covariance,n_iters,n_popn):

  # initialize the parameters

  dim=len(Returns.columns)

  lb=np.full(dim,0.05)
  ub=np.full(dim,1)

  individuals=np.zeros((n_popn,dim))
  fitness=np.zeros(n_popn)

  # generate random population and initial CSR

  for i in range(n_popn):
    individuals[i,:]=np.random.rand(1,dim)*(ub-lb)+lb

  new_indi=np.zeros((n_popn,dim))
  new_fit=np.zeros(n_popn)

  csr=np.zeros((n_popn,3))
  csr[:,0]=csr[:,1]=np.random.uniform(0,0.4)
  csr[:,2]=1-csr[:,0]-csr[:,1]

  count=np.zeros((n_popn,3))
  flag=np.zeros((n_popn,3))

#loop begins
  for t in range(n_iters):

    # evaluate fitness and rank in order

    for i in range(n_popn):
      # individuals[i,:]=spaceBound(individuals[i,:],lb,ub)
      fitness[i]=sharpeRatio(Returns,covariance,individuals[i,:])

    indF = np.argsort(fitness)[::-1]
    individuals = individuals[indF, :]
    fitness = fitness[indF]

    BestF = fitness[-1]
    BestX = individuals[-1, :]
    print(-BestF)
    # calculate population diversity and environmental stimulus

    Diversity = calculate_Diversity(individuals,n_popn,dim)
    stimulus = Diversity**(1.0/50.0)
##########################################################################################
    if stimulus>=1:

      r1=np.random.rand()
      a=(1-t/n_iters)*r1
      xrand = np.random.rand(dim)*(ub-lb)+lb
      new_indi[0,:]=(1-a)*individuals[n_popn-1,:]+a*xrand

      u = np.random.randn(1,dim)
      v = np.random.randn(1,dim)
      C = 0.5 * u / np.abs(v)

      new_indi[1,:] = individuals[1,:]+C*(individuals[1,:]-new_indi[0,:])

      for i in range(2,n_popn):

        maxi=np.argmin(csr[i,:])
        u = np.random.randn(1,dim)
        v = np.random.randn(1,dim)
        C = 0.5 * u / np.abs(v)

        if maxi==0:
          new_indi[i,:] = individuals[i,:]+C*(individuals[i,:]-new_indi[0,:])
        elif maxi==1:
          r2 = np.random.rand()
          rand_idx = np.random.randint(1,i)# [1,i)
          new_indi[i,:] = individuals[i,:]+C*(r2*(individuals[i,:]-new_indi[0,:])+(1-r2)*(individuals[i,:]-individuals[rand_idx,:]))
        else:
          rand_idx2 = np.random.randint(1,i)
          new_indi[i,:] = individuals[i,:]+C*(individuals[i,:]-individuals[rand_idx2,:])

        flag[i,maxi]=1

      for i in range(n_popn):
        new_indi[i,:] = spaceBound(new_indi[i,:],lb,ub)
        new_fit[i] = sharpeRatio(Returns,covariance,new_indi[i,:])

      for i in range(n_popn):
        for j in range(3):

          if flag[i,j]==1 and new_fit[i]<fitness[i]:
            fitness[i]=new_fit[i]
            individuals[i,:]=new_indi[i,:]
            count[i,j]=count[i,j]+1

          elif flag[i,j]==1 and new_fit[i]>=fitness[i]:
            for k in range(3):
              if k!=j:
                count[i,k]=count[i,k]+1

          csr[i,j]=count[i,j]/(2.0*(n_popn-1)*(t+1))
          flag[i,j]=0

#############################################################################################
    else :
      mini = np.argmin(fitness)

      for i in range(n_popn):

        r3 = np.random.rand()

        rand1=np.random.randint(0,i)

        rand2=np.random.randint(0,i)

        index = np.round(np.random.rand())+1

        D = 3*np.random.randn(1,dim)

        e = r3*(np.random.randint(1,3))-1

        h =2*r3-1

        #decomposition
        new_indi[i,:] = individuals[mini,:]+D*(e*individuals[mini,:]-h*individuals[i,:])+(individuals[rand1,:]+individuals[rand2,:])/2.0




  for i in range(n_popn):
    individuals[i,:]=spaceBound(individuals[i,:],lb,ub)
    fitness[i]=sharpeRatio(Returns,covariance,individuals[i,:])

  mini = np.argmin(fitness)
  BestF = fitness[mini]
  BestX = individuals[mini,:]

  return "optimal weights:",BestX/np.sum(BestX),-BestF



```

```python
def calculate_Diversity(individuals,n_popn,dim):
  outer_sum=0
  for i in range(n_popn):
    inner_sum=0
    for j in range(dim):
      inner_sum+=(individuals[i,j]-individuals.mean(axis=0)[j])**2
    outer_sum+=inner_sum

  return outer_sum/(n_popn*0.95)
```
