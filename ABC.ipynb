```python
def softmax(x):
    e_x = np.exp(x - np.max(x))
    return e_x / e_x.sum(axis=0)
```

```python
def sharpeRatio(Returns,covariance,unscaled_weights):

  weights=softmax(unscaled_weights)
  P_return=np.sum((Returns.mean()*weights))

  P_risk=np.sqrt(np.dot(weights.T,np.dot(covariance,weights)))

  return -P_return/P_risk
```

```python
def roulette_wheel_selection(P):
    r = np.random.rand()
    C = np.cumsum(P)
    i = np.argmax(r <= C)  # Find the first index where r <= C
    return i
```

```python
def ABC(Returns,covariance,n_iters,n_popn):

  dim=len(Returns.columns)
  n_onlookers=n_popn
  L=np.round(0.6*dim*n_popn)#Abandonment Limit Parameter (Trial Limit)
  a=1  #Acceleration Coefficient Upper Bound

  individuals=np.zeros((n_popn,dim))
  fitness=np.zeros(n_popn)

  for i in range(n_popn):
    individuals[i,:]=np.random.rand(1,dim)
    fitness[i]=sharpeRatio(Returns,covariance,individuals[i,:])

  indF = np.argsort(fitness)[::-1]
  individuals = individuals[indF, :]
  fitness = fitness[indF]

  BestF = fitness[-1]
  BestX = individuals[-1, :]

  C=np.zeros((n_popn,1))#Abandonment Counter

  new_indi=np.zeros((n_popn,dim))

  history=[]

  for it in range(n_iters):

    #Recruited bees
    for i in range(n_popn):

      # Choose k randomly, not equal to i
      K = list(range(0,i))+list(range(i + 1,n_popn))
      k = np.random.choice(K)

      # Define Acceleration Coeff.
      phi = a * np.random.uniform(-1, 1,dim)

      new_indi[i,:]=individuals[i,:]+phi*(individuals[i,:]-individuals[k,:])

      new_fit=sharpeRatio(Returns,covariance,new_indi[i,:])

      if new_fit<fitness[i]:
        individuals[i,:]=new_indi[i,:]
        fitness[i]=new_fit
      else:
        C[i]=C[i]+1

    #Calculate Fitness Values and Selection Probabilities
    meancost=np.mean(fitness)
    F=np.zeros(n_popn)
    for i in range(n_popn):
      F[i]=np.exp(-fitness[i]/meancost)
    P=F/np.sum(F)

    #Onlooker bees
    for m in range(n_onlookers):

      i=roulette_wheel_selection(P)

      # Choose k randomly, not equal to i
      K = list(range(0,i))+list(range(i + 1,n_popn))
      k = np.random.choice(K)

      # Define Acceleration Coeff.
      phi = a * np.random.uniform(-1, 1,dim)

      new_indi[i,:]=individuals[i,:]+phi*(individuals[i,:]-individuals[k,:])

      new_fit=sharpeRatio(Returns,covariance,new_indi[i,:])

      if new_fit<fitness[i]:
        individuals[i,:]=new_indi[i,:]
        fitness[i]=new_fit
      else:
        C[i]=C[i]+1

    #Scout Bees
    for i in range(n_popn):
      if C[i]>=L:
        individuals[i,:]=np.random.uniform(0,1,dim)
        fitness[i]=sharpeRatio(Returns,covariance,individuals[i,:])
        C[i]=0

    indF = np.argsort(fitness)[::-1]
    individuals = individuals[indF,:]
    fitness = fitness[indF]

    BestF = fitness[-1]
    BestX = individuals[-1,:]
    print(softmax(BestX))
    print(-BestF)
    history.append(-BestF)

  return history,softmax(BestX),-BestF

```

```python
def spaceBound(x,lb,ub):
  for i in range(len(x)):
    if x[i]<lb[i] or x[i]>ub[i]:
      x[i]=np.random.uniform(lb[i],ub[i])
  return x
```

```python
def AEO(Returns,covariance,n_iters,n_popn):
  #initialise an ecosystem, population of potential solutions

  dim=len(Returns.columns)

  lb=np.full(dim,0.00001) #lower bound
  ub=np.full(dim,1000) #upper bound

  #constraints add krna bacha hai

  #for each individual calculate fitness and store in array
  individuals=np.zeros((n_popn,dim))
  fitness=np.zeros(n_popn)


  for i in range(n_popn):

    individuals[i,:]=np.random.rand(1,dim)*(ub-lb)+lb

    fitness[i]=sharpeRatio(Returns,covariance,individuals[i,:])

  #sort fitness in increasing order of sharpe ratio and get the indexes
  indF = np.argsort(fitness)[::-1]

  # Reorder PopPos and PopFit based on sorted indices
  individuals = individuals[indF, :]
  fitness = fitness[indF]

  # Select the best fitness and corresponding position
  BestF = fitness[-1]
  BestX = individuals[-1, :]

  #n_iters for loop

  new_indi=np.zeros((n_popn,dim))

#loop1
  history_aeo=[]
  for it in range(n_iters):

    #production
    r1=np.random.rand()
    a=(1-it/n_iters)*r1
    xrand = np.random.rand(dim)*(ub-lb)+lb
    new_indi[0,:]=(1-a)*individuals[n_popn-1,:]+a*xrand #eq 1

    #first herbivore
    u = np.random.randn(1,dim)
    v = np.random.randn(1,dim)
    # Compute C based on the given equation
    C = 0.5 * u / np.abs(v)
    # Update new_indi based on the given equation
    new_indi[1,:] = individuals[1,:]+C*(individuals[1,:]-new_indi[0,:]) #eq 6

#loop2
    for i in range (2,n_popn):

      u = np.random.randn(1,dim)
      v = np.random.randn(1,dim)
      C = 0.5 * u / np.abs(v)#consumption factor
      r=np.random.rand()

      if r<1/3:
        new_indi[i,:] = individuals[i,:]+C*(individuals[i,:]-new_indi[0,:]) #eq 6

      elif r>2/3:
        r2 = np.random.rand()
        rand_idx = np.random.randint(1,i)# [1,i)
        new_indi[i,:] = individuals[i,:]+C*(r2*(individuals[i,:]-new_indi[0,:])+(1-r2)*(individuals[i,:]-individuals[rand_idx,:])) #eq 8

      else:
        rand_idx2 = np.random.randint(1,i)
        new_indi[i,:] = individuals[i,:]+C*(individuals[i,:]-individuals[rand_idx2,:]) #eq 7
#loop2 end

      new_fit=np.zeros(n_popn)
#loop3
    for i in range(n_popn):
      new_indi[i,:] = spaceBound(new_indi[i,:],lb,ub)
      new_fit[i] = sharpeRatio(Returns,covariance,new_indi[i,:])

      if new_fit[i]<fitness[i]:
        fitness[i]=new_fit[i]
        individuals[i,:]=new_indi[i,:]
#loop3 end

    mini = np.argmin(fitness)
#loop4
    for i in range(n_popn):
      r3 = np.random.rand()
      index = np.round(np.random.rand())+1
      D = 3*np.random.randn(1,dim)
      e = r3*(np.random.randint(1,3))-1
      h =2*r3-1
      new_indi[i,:] = individuals[mini,:]+D*(e*individuals[mini,:]-h*individuals[i,:]) #eq 9
#loop4 end

#loop5
    for i in range(n_popn):
      new_indi[i,:] = spaceBound(new_indi[i,:],lb,ub)
      new_fit[i] = sharpeRatio(Returns,covariance,new_indi[i,:])

      if new_fit[i]<fitness[i]:
        fitness[i]=new_fit[i]
        individuals[i,:]=new_indi[i,:]
#loop5 end

    indF = np.argsort(fitness)[::-1]
    individuals = individuals[indF, :]
    fitness = fitness[indF]

    if fitness[-1]<BestF:
      BestF = fitness[-1]
      BestX = individuals[-1, :]

    print(softmax(BestX),-BestF)
    history_aeo.append(-BestF)

  return history_aeo,softmax(BestX),-BestF
```
